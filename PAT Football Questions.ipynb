# Q5: # Q1. Create a new dataframe after removing the ourliers for Mins Column in the dataset (conside the upper quantile to be 99 percentile and the lower quantile to be 1 percentile.). After that Find the difference in the number of rows?

* Difference is df.shape(only rows) - df.shape(rows of filtered data only)

```python
lower = df['Mins'].quantile(0.01)
upper = df['Mins'].quantile(0.99)

new_df =  df[(df.Mins>=lower) & (df.Mins<=upper)]
difference_shape = df.shape[0] - new_df.shape[0]
print('Number of Rows/Observation before removing outliers : {}'.format(df.shape[0]))
print('Number of Rows/Observation after removing outliers : {}'.format(new_df.shape[0]))
print('Difference in Number of rows between them : {} '.format(difference_shape))
```

* Difference in the number of rows after removing outliers : 14
</div>

# Q6: Which top 5 players have scored the maximum number of goals throughout all the years? Display them in descending order

For this question we will be using loop to iterate every unique years that we have in our dataset

```python
player_scored_in_years = df.groupby(['Year','Player Names'])['Goals'].sum().reset_index()
top_scored_sorted_players = player_scored_in_years.sort_values(by=['Goals','Year'],ascending=[False,True])

for year in df.Year.unique():
    top_five_player_per_year = top_scored_sorted_players[top_scored_sorted_players.Year==year].head(5)
    print('At Year : {} , Top 5 Most scored players '.format(year))
    print(top_five_player_per_year.to_string(index=False))
    print()
```

<div class="alert alert-block alert-info">
<b>Key Findings: </b> 
    
    
* Positive Correlation: 
    
    - There is a constant relationship between the total number of shots taken in a match and the number of shots that are on target, as seen by the high and positive correlation between these two variables. To put it another way, teams or players that attempt more shots throughout a game also typically have more shots on goal.
    
    - More minutes played by players often translate into more shots being taken. Players who play longer stretches of time on the pitch are more likely to try shots on goal, thus this association makes logical.
    
    - when players generate better goal-scoring opportunities, they are more likely to convert those opportunities into goals, resulting in a positive correlation.
    
    
* Negative Correlation:
    - A negative correlation implies that the number of matches played tends to decline as the years go by (increase). Stated differently, there is a tendency towards players or teams playing in fewer games over time.

# Q 13. Which Player has the least and most number of missed shots in the dataset?
* Missed Shots=Shots-On Target

```python
df_temp = df.copy()
df_temp['Missed_Shots'] = df_temp['Shots'] - df_temp['OnTarget'] # Missed Shots = (Shots - On Target)
least_missed_shots = df_temp[df_temp['Missed_Shots'] == df_temp['Missed_Shots'].min()]
most_missed_shots = df_temp[df_temp['Missed_Shots'] == df_temp['Missed_Shots'].max()]

print("Player with the Least Missed Shots:")
print(least_missed_shots[['Player Names', 'Missed_Shots']])
print("\nPlayer with the Most Missed Shots:")
print(most_missed_shots[['Player Names', 'Missed_Shots']])
```

# Q15: Draw a Line Chart to show the comparison of the top 5 highest scoring players in the dataset throughout the years(2016-2020)

```python
years = list(range(2016, 2021))
year_filter = df[df['Year'].isin(years)]
player_goals = year_filter.groupby(['Player Names', 'Year'])['Goals'].sum().reset_index()
top_players = player_goals.groupby('Player Names')['Goals'].sum().nlargest(5).index
top_players_df = player_goals[player_goals['Player Names'].isin(top_players)]
plt.figure(figsize=(10, 6))

for player in top_players:
    player_data = top_players_df[top_players_df['Player Names'] == player]
    plt.plot(player_data['Year'], player_data['Goals'], marker='o', label=player)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Goals', fontsize=12)
plt.title('Comparison of the Top 5 Highest-Scoring Players (2016-2020)', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
